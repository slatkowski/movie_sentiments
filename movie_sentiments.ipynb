```python
df = pd.read_csv("reviews.csv")
df
```

```python
for review in df['review']:
    sents = sent_tokenize(review)
    words = word_tokenize(review)

    print("Zdania:")
    for sent in sents:
        print(sent)

    print("Słowa:")
    for word in words:
        print(word)
```
