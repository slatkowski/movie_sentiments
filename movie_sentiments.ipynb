```python
df = pd.read_csv("reviews.csv")
df
```

```python
df['review'] = df['review'].str.lower()
```

```python
sents = []
words = []

for review in df['review']:
    sents.append(sent_tokenize(review))
    words.append(word_tokenize(review))
```

```python
stopwords = ""

with open("stopwrods.txt", "r", encoding = "utf-8") as file:
    stopwords = file.read()

stopwords = stopwords.splitlines()
stopwords
```

```python
def remove_stopwords(x: str) -> str:
    return ", ".join([word for word in str(x).split() if word not in stopwords])

df['review'] = df['review'].apply(lambda x: remove_stopwords(x))
```
